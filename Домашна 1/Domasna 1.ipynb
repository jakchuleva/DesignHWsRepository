```python
snapshot = requests.get(snapshot_url)
```

```python
html = BeautifulSoup(snapshot.text, 'html.parser')
```

```python
html
```

```python
table = html.find('table', {'class': "wikitable"})
```

```python
table
```

```python
df=pd.read_html(str(table))
# convert list to dataframe
df=pd.DataFrame(df[0])
print(df.head())
```
